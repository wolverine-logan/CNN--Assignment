```python
def predict(data_file, weights):
    data_file_matrix  = data_file.as_matrix()[:,:-1]
    import numpy as np
    bias_multiplier =  np.ones(len(data_file_1))
    data_file_matrix = np.c_[data_file_matrix, bias_multiplier]
    activation = np.dot(data_file_matrix, weights)
    y_predicted = []
    for value in activation:
        if value > 0 :
            y_predicted.append(1)
        else:
            y_predicted.append(0)
    return y_predicted
```

```python
weights = np.random.random_sample(3)
```

```python
def train(data_file, epoch_counts, weights, learning_rate):
    y_observed = np.array(data_file.loc[:,'y'])
    data_file_matrix  = data_file.as_matrix()[:,:-1]
    bias_multiplier =  np.ones(len(data_file_1))
    data_file_matrix = np.c_[data_file_matrix, bias_multiplier]
    weight = []
    for epoch in range(0,epoch_counts):
        total_error = 0.0
        k = 0
        for row in data_file_matrix:
            y_predict = np.dot(row,weights)
            if y_predict > 0:
                y_predict = 1
            else:
                y_predict = 0
            error = y_observed[k] - y_predict
            k += 1
            total_error += error**2
            weights = weights + learning_rate * error * row
        print('epoch=',epoch, '\tlrate=',learning_rate,'\terror=',total_error)
        if(total_error == 0.0):
            break
        weight.append(weights)
    return weight
        
    
```

```python
weights_1 = train(data_file_1,40,weights,0.1)
```

```python
weights_2 = train(data_file_2,40,weights,0.1)
```

```python
weights_3 = train(data_file_3,40,weights,0.1)
```

```python
weights_4 = train(data_file_4,40,weights,0.1)
```

```python
def line_func(x, weights):
    y = -((weights[-1][0]/weights[-1][1])*x - (weights[-1][2]/weights[-1][1]))
    return y
```

# Animation Plots

```python
# First set up the figure, the axis, and the plot element we want to animate
%matplotlib nbagg
fig = plt.figure()
ax = plt.axes(xlim=(-2, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)
x1 = data_file_1['x1']
x2 = data_file_1['x2']
y = data_file_1['y']
w = weights_1
plt.scatter(x = data_file_1[data_file_1['y']==0].x1, y = data_file_1[data_file_1['y']==0].x2, edgecolors='g', marker = 'X', label='y=0')
plt.scatter(x = data_file_1[data_file_1['y']==1].x1, y = data_file_1[data_file_1['y']==1].x2, edgecolors='m', marker = 'o', label='y=1')


# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = x1
    y = -((w[i][0]/w[i][1])*x - (w[i][2]/w[i][1]))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=len(weights_2), interval=200, blit=True, repeat = False)

plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(['hyperline','y=0','y=1'])
plt.title('x1 vs x2')
plt.show()
```
